[剑指 Offer 37. 序列化二叉树](https://leetcode-cn.com/problems/xu-lie-hua-er-cha-shu-lcof/)

```
请实现两个函数，分别用来序列化和反序列化二叉树。

示例: 

你可以将以下二叉树：

    1
   / \
  2   3
     / \
    4   5

序列化为 "[1,2,3,null,null,4,5]"


```

````python
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, x):
#         self.val = x
#         self.left = None
#         self.right = None

class Codec:

    def serialize(self, root):
        """Encodes a tree to a single string.
        
        :type root: TreeNode
        :rtype: str
        """
        if root == None:
            return "[]"
        queue = []
        reocrd = []
        r = root
        queue.append(r)

        
        while len(queue):
            tmp = queue[0]
            queue.pop(0)
            reocrd.append(str(tmp.val) if tmp else "null")
            if tmp:
                queue.append(tmp.left)
                queue.append(tmp.right)


        cur = 0
        for i in range(len(reocrd)-1,-1,-1):
            if reocrd[i] != "null":
                cur = i
                break

        re = ""
        for i in range(cur+1):
            re += reocrd[i]
            if i!=cur:
                re += ','

        re = "[" + re + "]"
        return re


    def deserialize(self, data):
        """Decodes your encoded data to tree.
        
        :type data: str
        :rtype: TreeNode
        """
        if data=="[]":
            return None

        data = data[1:-1]
        fields = data.split(",")
        head = TreeNode(fields[0])
        queue = []
        queue.append(head)

        i = 1
        while len(queue):
            tmp = queue[0]
            queue.pop(0)
            if i<len(fields) and fields[i]!="null":
                l = TreeNode(fields[i])
                tmp.left = l
                queue.append(l) 
            i+=1 
            if i<len(fields) and fields[i]!="null":
                r = TreeNode(fields[i])
                tmp.right = r
                queue.append(r)
            i+=1

        return head

# Your Codec object will be instantiated and called as such:
# codec = Codec()
# codec.deserialize(codec.serialize(root))
````
